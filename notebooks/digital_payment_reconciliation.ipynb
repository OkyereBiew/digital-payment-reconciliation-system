```python
# =========================================
# DIGITAL PAYMENT RECONCILIATION SYSTEM
# PHASE 1 — DATA LOADING & INSPECTION
# =========================================

# Import libraries
import pandas as pd
import numpy as np

# Display settings
pd.set_option('display.max_columns', None)

# Load dataset
df = pd.read_csv("/onlinefraud.csv")

# View first rows
print("FIRST 5 ROWS")
display(df.head())

# Dataset shape
print("\nDATASET SHAPE")
print(df.shape)

# Column names
print("\nCOLUMN NAMES")
print(df.columns)

# Missing values
print("\nMISSING VALUES")
print(df.isnull().sum())
```
